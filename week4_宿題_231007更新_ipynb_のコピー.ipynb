```python
# working place. everything
def homework(target_online_retail_data_tb, n):
    df1 = target_online_retail_data_tb.groupby('CustomerID', as_index = False)['TotalPrice'].sum()
    df2 = df1.sort_values(by='TotalPrice', ascending=False)
    TotalPrice_sum = df2['TotalPrice'].sum()
    bins = pd.qcut(df2.TotalPrice, n)
    df3 = pd.DataFrame({'Data': df2.TotalPrice, 'Quantile': bins})
    Quantile_sum = df3.groupby('Quantile', as_index=False)['Data'].sum()
    Quantile_propotion = Quantile_sum['Data']/TotalPrice_sum
    my_result = Quantile_propotion.sort_values(ascending = False)
    return my_result
```
